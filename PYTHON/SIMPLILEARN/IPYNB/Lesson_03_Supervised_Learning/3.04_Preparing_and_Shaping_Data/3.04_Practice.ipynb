```python
'''
Here's a breakdown of the columns:

        PassengerId: Unique identifier for each passenger.
           Survived: Indicates whether the passenger survived (1) or not (0).
             Pclass: Ticket class (1st, 2nd, or 3rd).
               Name: Name of the passenger.
                Sex: Gender of the passenger.
                Age: Age of the passenger. (Note: There are missing values, indicated by the difference in non-null counts)
              SibSp: Number of siblings/spouses aboard the Titanic.
              Parch: Number of parents/children aboard the Titanic.
             Ticket: Ticket number.
               Fare: Passenger fare.
              Cabin: Cabin number. (Note: There are missing values)
           Embarked: Port of embarkation (C = Cherbourg, Q = Queenstown, S = Southampton). (Note: There are missing values)
'''
```

```python
'''
1. Create a new column called travel_alone
2. Check if Sibling/spouse + Parent/child > 0
3. If it is greater than 0, then assign 0 to the travel_alone or else assign 1.

NOTE: USE astype('uint8') ---> converts resulting boolean values to unsigned integer type of 8 bits.
'''

# dat_fr['travel_alone'] = np.where((dat_fr['SibSp'] + dat_fr['Parch']) > 0, 0, 1).astype('uint8')
#                                                                         |  |  |
#                                                                         a  b  c
# a -> condition
# b -> if true then assign value b
# c -> if false then assign value c

dat_fr['travel_alone'] = np.where( ( dat_fr['SibSp'] + dat_fr['Parch'] ) > 0, 0, 1 )
```

```python
dat_fr
```

```python
dat_fr.info()
```

```python
# Dropping Unwanted Columns
dat_fr1 = dat_fr.drop( ['PassengerId', 'Name', 'SibSp', 'Parch', 'Ticket', 'Cabin'], axis=1 )
```

```python
dat_fr1
```

```python
# Checking for missing values - Age, Cabin has missing values
dat_fr.isna().sum()
```

```python
# Checking for missing values - Age has missing values
dat_fr.isna().sum()
```

```python
dat_fr['Age'].fillna( dat_fr1['Age'].median( skipna=True ), inplace=True )
```

```python
# We can see that the Age column now does not have any null values in dat_fr
dat_fr.isna().sum()
```

```python
dat_fr1.head()
```

```python
# Creating dummy variables, basically ---> ONE HOT ENCODING
dat_fr_titanic = pd.get_dummies(dat_fr1, columns=['Pclass', 'Embarked', 'Sex'], drop_first=True)
```

```python
dat_fr_titanic.head()
```

```python
dat_fr_titanic = dat_fr_titanic.replace( {True: 1, False: 0} )
```

```python
dat_fr_titanic.head()
```

```python
# Creating Scalars
X = dat_fr_titanic.drop(['Survived'], axis=1)
Y = dat_fr_titanic['Survived']
```

```python
X
```

```python
Y
```

```python
# PREPROCESSING

from sklearn.preprocessing import MinMaxScaler, StandardScaler
```

```python
# Transforming

trans_MM = MinMaxScaler()
trans_SS = StandardScaler()

dat_fr_MM = trans_MM.fit_transform(X)
pd.DataFrame(dat_fr_MM)


'''
We can see the data for Age and Fare to be in the range of [0 to 1] -----> MINMAXSCALER
'''
```

```python
dat_fr_SS = trans_SS.fit_transform(X)
pd.DataFrame(dat_fr_SS)

'''
We can see the data for Age and Fare to be in the range of [-1 to 1] -----> STANDARDSCALER
'''
```
